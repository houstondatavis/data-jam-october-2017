```python
df_g_beats
```

```python
df['Beats'] = df.district_beat
df_g_mergeAll = df_g_beats.merge(df, on='Beats')
df_g_mergeAll
```

```python
### DON"T DO THIS TAKES LONG TIME
df_g_mergeAll.plot()
```

```python
kw = dict(column='age', k=6, colormap='YlGn')
ax = df_g_mergeAll.plot(scheme='QUANTILES', **kw)
```

```python
columns = ['age','sex']
df_g_some = df_g_mergeAll[columns]
kw = dict(column='age', k=6, colormap='YlGn')
ax = df_g_some.plot(scheme='equal_interval', **kw)
```

```python
kw = dict(column='age', k=6, colormap='YlGn')
ax = df_g_mergeAll[df_g_mergeAll['Beats']=='1A10'].plot(scheme='equal_interval', **kw)
```

```python
# list of top 10 beats = list_top_beats
kw = dict(column='age', k=6, colormap='YlGn')
for beat in list_top_beats:
    ax = df_g_mergeAll[df_g_mergeAll['Beats']==beat].plot(scheme='equal_interval', **kw)
```
